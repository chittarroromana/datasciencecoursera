```
from __future__ import print_function

from apiclient import discovery
from httplib2 import Http
from oauth2client import file, client, tools

TMPLFILE = 'Template 2 Deck'   # use your own!
SCOPES = (
    'https://www.googleapis.com/auth/drive',
    'https://www.googleapis.com/auth/presentations',
    'https://www.googleapis.com/auth/spreadsheets',
)
store = file.Storage('storage.json')
creds = store.get()
if not creds or creds.invalid:
    flow = client.flow_from_clientsecrets('client_id.json', SCOPES)
    creds = tools.run_flow(flow, store)
HTTP = creds.authorize(Http())
DRIVE  = discovery.build('drive',  'v3', http=HTTP)
SLIDES = discovery.build('slides', 'v1', http=HTTP)
sheets_service = discovery.build('sheets', 'v4', http=HTTP)

# Use the Sheets API to load data, one record per row.
data_range_notation = 'Surreal!A2:Z999'
sheets_response = sheets_service.spreadsheets().values().get(
    spreadsheetId='1YBe2ef7Sxktswylcl0obZaDNEy-r97tX5330gX9uLCQ',
    range=data_range_notation).execute()
values = sheets_response.get('values')
print(values)
#i=1; j=3; print(values[i][j])

###############################################################################
```

```
template_presentation_name = input("Enter the name of the template presentation")

version = input("Version number?")

rsp = DRIVE.files().list(q="name='%s'" % template_presentation_name).execute().get('files')[0]

template_presentation_id = DRIVE.files().copy(fileId=rsp['id']).execute().get('id')

# For each record, create a new merged presentation.
for row in values:
    style = row[0]       # style of the template
    title = row[1]    # title of the new presentation
    img1 = row[2]
    img2 = row[3]
    img3 = row[4]
    hook1 = row[5]
    hook2 = row[6]
    hook3 = row[7]
    hook4 = row[8]
    hook5 = row[9]
    boom1 = row[10]
    theme = row[11]
    themeQ = row[12]
    xlimg1 = row[13]
    # Duplicate the template presentation using the Drive API.
    copy_title = 'v' + version + ' ' + style + ' ' + title + ' presentation'
    body = {
        'name': copy_title
    }
    drive_response = DRIVE.files().copy(
        fileId=template_presentation_id, body=body).execute()
    presentation_copy_id = drive_response.get('id')

    reqs = [
    {'replaceAllText': {
        'containsText': {'text': '{{TITLE}}'},
        'replaceText': title
    }},
    {'replaceAllText': {
        'containsText': {'text': '{{hook1}}'},
        'replaceText': hook1
    }},
    {'replaceAllText': {
        'containsText': {'text': '{{hook2}}'},
        'replaceText': hook2
    }},
    {'replaceAllText': {
        'containsText': {'text': '{{hook3}}'},
        'replaceText': hook3
    }},
    {'replaceAllText': {
        'containsText': {'text': '{{hook4}}'},
        'replaceText': hook4
    }},
    {'replaceAllText': {
        'containsText': {'text': '{{hook5}}'},
        'replaceText': hook5
    }},
    {'replaceAllText': {
        'containsText': {'text': '{{boom1}}'},
        'replaceText': boom1
    }},
    {'replaceAllText': {
        'containsText': {'text': '{{theme}}'},
        'replaceText': theme
    }},
     {'replaceAllText': {
        'containsText': {'text': '{{themeQ}}'},
        'replaceText': themeQ
    }},
   ]

    # Execute the requests for this presentation.
    body = {
        'requests': reqs
    }
    response = SLIDES.presentations().batchUpdate(
        presentationId=presentation_copy_id, body=body).execute()
    # Count the total number of replacements made.
    num_replacements = 0
    for reply in response.get('replies'):
        num_replacements += reply.get('replaceAllText') \
            .get('occurrencesChanged')
    print('Created presentation for %s with ID: %s' %
        (title, presentation_copy_id))
    print('Replaced %d text instances' % num_replacements)

    requests = []
    requests.append({
        'replaceAllShapesWithImage': {
            'imageUrl': img1,
            #'replaceMethod': 'CENTER_INSIDE',
            'containsText': {
                'text': '{{themeimg1}}',
                #'matchCase': True
            }
        }
    })
    requests.append({
        'replaceAllShapesWithImage': {
            'imageUrl': img2,
          #  'replaceMethod': 'CENTER_INSIDE',
            'containsText': {
                'text': '{{themeimg2}}',
                #'matchCase': True
            }
        }
    })

    requests.append({
        'replaceAllShapesWithImage': {
            'imageUrl': img3,
            #'replaceMethod': 'CENTER_INSIDE',
            'containsText': {
                'text': '{{titleimg}}',
                #'matchCase': True
            }
        }
    })

    # Execute the requests.
    body = {
        'requests': requests
    }
    response = SLIDES.presentations().batchUpdate(
        presentationId=presentation_copy_id, body=body).execute()

    # Count the number of replacements made.
    num_replacements = 0
    for reply in response.get('replies'):
       num_replacements += reply.get('replaceAllShapesWithImage') \
           .get('occurrencesChanged')
    print('Created merged presentation with ID: {0}' \
        .format(presentation_copy_id))
    print('Replaced %d shapes with images.' % num_replacements)
```
